# 01 EDA

KOBIS/TMDB 매칭 데이터의 관객 수 분포, 클래스 비율, 장르별 관객 수를 확인합니다.

```python
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt

ROOT = Path.cwd()
if ROOT.name == 'notebooks':
    ROOT = ROOT.parent

data = pd.read_csv(ROOT / 'data/kobis_tmdb_title_matches.csv', encoding='utf-8-sig')
data['audience_count'] = pd.to_numeric(data['audience_count'], errors='coerce')
data.shape
```

```python
data[['match_title', 'open_date', 'audience_count', 'tmdb_genres']].head()
```

```python
data['is_success_3000000'] = data['audience_count'] >= 3_000_000
fig, axes = plt.subplots(1, 2, figsize=(12, 4))
data['audience_count'].plot(kind='hist', bins=30, ax=axes[0], title='Audience Count')
data['is_success_3000000'].value_counts().sort_index().plot(kind='bar', ax=axes[1], title='Success Class')
plt.tight_layout()
```

```python
genre_rows = []
for _, row in data.dropna(subset=['tmdb_genres']).iterrows():
    for genre in str(row['tmdb_genres']).split('|'):
        if genre:
            genre_rows.append({'genre': genre, 'audience_count': row['audience_count']})
genre_summary = pd.DataFrame(genre_rows).groupby('genre')['audience_count'].agg(['count', 'mean']).sort_values('mean', ascending=False)
genre_summary.head(15)
```
