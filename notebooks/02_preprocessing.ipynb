```python
from pathlib import Path
import subprocess
import sys

ROOT = Path.cwd()
if ROOT.name == 'notebooks':
    ROOT = ROOT.parent

python = sys.executable
subprocess.run([
    python,
    str(ROOT / 'src/feature_poster.py'),
    '--input', str(ROOT / 'data/kobis_tmdb_title_matches.csv'),
    '--poster-dir', str(ROOT / 'data/posters'),
    '--output', str(ROOT / 'data/processed/poster_features.csv'),
], check=True, cwd=ROOT)
subprocess.run([
    python,
    str(ROOT / 'src/preprocessing.py'),
    '--input', str(ROOT / 'data/kobis_tmdb_title_matches.csv'),
    '--output', str(ROOT / 'data/processed/movie_features.csv'),
], check=True, cwd=ROOT)
```
